```python
# from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.random.randint(0,60,100)
```

```python
y = np.random.randint(40,100,100)
```

```python
x = x.reshape(-1,1)
# x
```

```python
y = y.reshape(-1,1)
```

```python
plt.hist(x)
plt.hist(y)
```

```python
x_scaler = MinMaxScaler()
x_minmax = x_scaler.fit_transform(x)
plt.hist(x_minmax)
```

```python
y_scaler = MinMaxScaler()
y_minmax = y_scaler.fit_transform(y)
plt.hist(y_minmax)
```

```python
plt.hist(x_minmax,bins=20)
plt.hist(y_minmax,bins=20)
```

```python
# x_minmax,y_minmax
```

```python
x_scaler = StandardScaler()
x_minmax = x_scaler.fit_transform(x)
plt.hist(x_minmax)
```

```python
y_scaler = StandardScaler()
y_minmax = y_scaler.fit_transform(y)
plt.hist(y_minmax)
```

```python
plt.hist(x_minmax,bins=20)
plt.hist(y_minmax,bins=20)
```
