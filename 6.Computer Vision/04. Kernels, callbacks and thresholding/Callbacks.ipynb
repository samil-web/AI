```python
# Load an image and show it on screen with cv2.imshow
# When the key g or G is detected then change the image to show in grayscale
# When the key c or C is detected then change the image to show in color
# When the key t or T is detected then change the image to show the thresholded version of the image
# When the key esc is detected close the image

shamil = cv2.imread('img/download.jfif')
shamil_copy = shamil.copy()
key = 0

# cv2.waitKey(0)


while(True):
    if (key == ord('g')) or (key == ord('G')):
        gray = cv2.cvtColor(shamil_copy,cv2.COLOR_BGR2GRAY)
        cv2.imshow('gray',gray)
    if (key == ord('c')) or (key == ord('C')) :
        color = cv2.cvtColor(shamil_copy,cv2.COLOR_BGR2RGB)
        cv2.imshow('color',color)
    if (key == ord('t')) or (key == ord('T')) :
        ret,thresh = cv2.threshold(shamil_copy,100,255,cv2.THRESH_BINARY)
        cv2.imshow('thresh',thresh)
    if key == 27: #27 represents ascii code for escape button
        break
    cv2.imshow('image',shamil_copy)

    key = cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Load an image and show it on screen with cv2.imshow
# When the key g or G is detected then change the image to show in grayscale
# When the key c or C is detected then change the image to show in color
# When the key t or T is detected then change the image to show the thresholded version of the image
# When the key esc is detected close the image

count = 0


cap = cv2.VideoCapture(0)
key = 0

img.shape
img_def = np.ones(img.shape,dtype='uint8')
img_def.shape
while(True):
    ret,frame = cap.read()

    if key == ord('q'):
        break
    if key == ord('i'):
        count += 1
        img_def = frame.copy()
        cv2.putText(img_def,f'{count}',(200,200),cv2.FONT_HERSHEY_COMPLEX,2,(0,255,0),2)
    cv2.imshow('Img',img_def)
    if key == ord('d'):
        count += 1
        img_def = frame.copy()
        cv2.putText(img,f'{count}',(200,200),cv2.FONT_HERSHEY_COMPLEX,2,(0,255,0),2)
    cv2.imshow('Img',img_def)

    key = cv2.waitKey(0)
print(ret)
cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```

## Mouse callbacks

```python
# Load an image and show it on screen with cv2.imshow
# When the user clicks on top of the image save the coordinates
# The user will then move the cursor without releasing the mouse button to another position
# When the user releases the button draw a rectagle around the ROI
# When the key esc is detected close the image
window_name = 'First Window'
load = cv2.imread('img/shamil.jpg')
copy = load.copy()
# cv2.imshow('copy',copy)
cv2.namedWindow(window_name)
key = 0

def right_left_click(action,x,y,flags,userdata):
    if action == cv2.EVENT_LBUTTONDOWN:
        cv2.rectangle(userdata,(x,y),(x+50,y+50),(0,0,155),cv2.FILLED)
        cv2.imshow(window_name,userdata)
    if action ==cv2.EVENT_RBUTTONDOWN:
        cv2.rectangle(userdata,(x,y),(x+50,y+50),(0,0,0),cv2.FILLED)
        cv2.imshow(window_name,userdata)
        
    if key == 27:
        cv2.destroyAllWindows()


cv2.setMouseCallback(window_name,right_left_click,copy)
cv2.imshow(window_name,copy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
