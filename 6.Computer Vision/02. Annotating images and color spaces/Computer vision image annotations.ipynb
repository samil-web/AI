### To edit an image making a copy of it is always a good idea!

```python
# Load an image
# And make a copy of the original image to avoid manipulating the original image
blue_flowers = cv2.imread('img/blue-flowers.jpg')
```

```python
plt.figure(figsize=(20,15))
plt.imshow(blue_flowers)
```

```python
copied = blue_flowers.copy()
```

[Hint](# "Numpy .copy function") 

## Drawing

### Drawing lines

```python
# Draw some lines on top of the image 
# copied
green  = (0,255,0)
new_img = cv2.line(copied,(10,10),(50,50),green,3)
plt.imshow(new_img)
```

[Hint](# "cv2.line function") 

### Circles

```python
# Draw some circles on top of the image
copied = blue_flowers.copy()
circle = cv2.circle(copied,(50,50),50,green,3)
plt.figure(figsize=(20,15))
plt.imshow(circle)
```

[Hint](# "cv2.circle") 

### Ellipses

```python
# Draw some elipses on top of the image
copied = blue_flowers.copy()

ellipse = cv2.ellipse(copied,(100,100),(50,50),90,0,360,green,2)
plt.figure(figsize=(20,20))
plt.imshow(ellipse)
```

[Hint](# "cv2.ellipse") 

### Poligons

```python
# Lets draw some poligons, first declare the points of the poligon
points = np.array([[10,10],[244,9],[44,90],[60,60]])
```

[Hint](# "Example: np.array([[10,5],[120,130],[170,120],[250,210]], np.int32)") 

```python
# Draw the poligon on top of the image
copied = blue_flowers.copy()
# points = np.array([[10,10],[244,9],[44,90],[60,60]])
polygon = cv2.polylines(copied,[points],True,(0,0,255))
plt.figure(figsize=(20,20))
plt.imshow(copied)
```

### Add an overlay on top of an image

```python
# Load an image and create a copy of it you can call it overlay
# Create a rectangle with a fill color (-1) on top of the overlay image
# Combine both images setting some opacity on the overlay
# Add some text on top of the overlay and show the image on screen
overlay = blue_flowers.copy()
cv_img = cv2.imread('img/purple-flowers.jpg')
overlay2 = cv_img.copy()
cv2.rectangle(overlay,(59,59),(40,40),(100,200,200),-1)
# added_img = cv2.addWeighted(overlay,1,overlay2,0.1,1)
plt.imshow(overlay2)
plt.figure(figsize=(20,20))
print(overlay2.shape)
overlay.shape
# plt.imshow(added_img)
```
