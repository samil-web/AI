## Histograms

```python
# Load some images, calculate and plot their color histograms, one color at the time
# try it with different color spaces BGR, HSV, etc
load_img = cv2.imread('img/noisy.png')
plt.imshow(load_img)
```

```python
cv2.imshow('img',load_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# help(plt.subplots)
```

```python
# calcHist
colors = ['b','g','r']
for i,color in enumerate(colors):
    histogram = cv2.calcHist([load_img],[i],None,[255],[0,255])
    plt.plot(histogram,color=color)
    plt.xlim(0,256)
    # fig,(ax1,ax2) = plt.subplots(1,2,1)
    # ax1.plot(histogram,load_img)
    # ax2.plot(histogram,load_img*(-1))
    # histogram,load_img = plt.subplots(1,2)
```
